```python
import networkx as nx 
nx.__version__
```

```python
#plt.figure(figsize=(10,10))
pos_forceatlas2 = nx.forceatlas2_layout(generator.gpx,max_iter=1000,scaling_ratio=10,linlog=False)
#scaling_ratio float (default: 2.0) Determines the scaling of attraction and repulsion forces.
# linlog
# bool (default: False)
# Uses logarithmic attraction instead of linear.

pos_spring = nx.spring_layout(generator.gpx,iterations=1000,method="energy", scale=1)

joint_array = np.vstack(list(pos_forceatlas2.values()))
joint_array_forceatlas2 = nx.rescale_layout(joint_array,scale=generator.size/2)+generator.size/2
del joint_array

joint_array = np.vstack(list(pos_spring.values()))
joint_array_spring = nx.rescale_layout(joint_array,scale=generator.size/2)+generator.size/2

plt.plot(joint_array_forceatlas2[:,0],joint_array_forceatlas2[:,1],'o',label='ForceAtlas2')
plt.plot(joint_array_spring[:,0],joint_array_spring[:,1],'o',label='Spring')
plt.legend()
plt.show()
```
