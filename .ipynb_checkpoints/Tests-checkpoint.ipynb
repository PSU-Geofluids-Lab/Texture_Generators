```python
from scipy import stats
from skimage.morphology import disk
import pandas as pd

oe.reload(mtr)
im = texture_data.copy()
num_steps = 5
x_all = np.zeros([num_steps,100])
y_all = np.zeros([num_steps,100])

for i in range(num_steps):
    snow,im_use = mtr.get_regions_segment(im,sigma=0.5,r_max=10,max_phase=2,filepath=None,print_stats=False)
    if i == 0:
        df_prop_summary_all = mtr.make_partition_regionprop(im_use,snow.regions,
                                                        plot_specific_region=False,region_id=10,
                                                        summary_images=True,print_Stats=False)
        df_prop_summary_all['label'] = i
    else :
        df_prop_summary = mtr.make_partition_regionprop(im_use,snow.regions,
                                                        plot_specific_region=False,region_id=10,
                                                        summary_images=True,print_Stats=False)
        df_prop_summary['label'] = i
        df_prop_summary_all = pd.concat([df_prop_summary_all,df_prop_summary])
    im = ps.filters.fftmorphology(im, strel=disk(1), mode='erosion')
                                         
plt.hist2d(df_prop_summary_all['label'].values,np.log10(df_prop_summary_all['area'].values), bins=[num_steps,10],cmap='RdYlBu_r')
plt.xlabel('Evolution Step')
plt.ylabel('Area Distribution')
plt.colorbar(label='Count')
plt.show()
```
